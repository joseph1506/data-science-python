```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x= np.arange(5,10,0.1)
```

```python
y= 2*x +1
y1=5*x+2
```

```python
fig=plt.figure(figsize=(10,5))
plt.plot(x,y,linewidth=2.0,linestyle=':',color='blue',marker='o',alpha=0.5)
plt.plot(x,y1,linewidth=2.0,linestyle=':',color='red',marker='o',alpha=0.5)
plt.title('Line Plot Demo')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend(['2XlINE','5xlINE'])
plt.grid(True)
plt.show()
```

```python
data ={"mangoes":20,"apples":30,"lemon":13,"oranges":90}
```

```python
names = list(data.keys())
values = list(data.values())
```

```python
fig=plt.figure(figsize=(10,5))
```

```python
plt.bar(names,values)
plt.show()
```

```python
plt.barh(names,values,color="orange")
plt.title('Horizontal bar')
plt.xlabel('Quantity')
plt.ylabel('Fruits')
plt.savefig('Barh.png')
plt.show()
```
